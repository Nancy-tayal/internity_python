# Matplotlib Styles 
With matplotlib, we can style the plots like, an HTML webpage is styled by using CSS styles. We just need to import style package of matplotlib library. 

There are various built-in styles in style package, and we can also write customized style files and, then, to use those styles all you need to import them and apply on the graphs and plots.

```python
#to find all the available styles use
from matplotlib import style, pyplot as plt
print(plt.style.available)
```

```python
import numpy as np
data = np.random.randn(50)
plt.style.use('bmh')
plt.plot(data)
plt.show()

plt.style.use('grayscale')
plt.plot(data)
plt.show()

plt.style.use('ggplot')
plt.plot(data)
plt.show()
```

```python
data = np.random.randn(50)
plt.style.use(['grayscale','dark_background']) #styles combined by passing a list of styles
#Note that styles further to the right will overwrite values that are already defined by styles on the left.
plt.plot(data)
plt.show()
```

# Temporary Styling
If you only want to use a style for a specific block of code but don't want to change the global styling, the style package provides a context manager for limiting your changes to a specific scope. To isolate your styling changes, you can write something like the following:

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

# Subplot:

The Matplotlib subplot() function can be called to plot two or more plots in one figure. Matplotlib supports all kind of subplots including 2x1 vertical, 2x1 horizontal or a 2x2 grid
Using subplot() method. 
This method adds another plot to the current figure at the specified grid position.

```python
import matplotlib.pyplot as plt
x = [3, 1, 3]
y = [3, 2, 1]
z = [1, 3, 1]
plt.figure()
plt.subplot(221)
plt.plot(x, y)

plt.subplot(222)
plt.plot(z, y)
plt.plot(z,x)
plt.show()

plt.subplot(223)
plt.plot(y,z)

plt.subplot(224)
plt.plot(y, x)
plt.show()
```

# Subplots:
Using subplots() method
This function is used to create figure and multiple subplots at the same time.

```python
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use("bmh")
# Creating the figure and subplots
# according the argument passed
fig, axes = plt.subplots(1, 2)

# plotting the data in the 1st subplot
axes[0].plot([1, 2, 3, 4], [1, 2, 3, 4])

# plotting the data in the 1st subplot only
axes[0].plot([1, 2, 3, 4], [4, 3, 2, 1])

# plotting the data in the 2nd subplot only
axes[1].plot([1, 2, 3, 4], [1, 1, 1, 1])
plt.show()
```

# Multiple charts, figures and subplots

Using Subplots

```python
# importing libraries
import numpy as np
import math

X = np.arange(0, math.pi*2, 0.05)

Y1 = np.sin(X)
Y2 = np.cos(X)
Y3 = np.tan(X)
Y4 = np.tanh(X)

# Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(2, 2)

# For Sine Function
axis[0, 0].plot(X, Y1)
axis[0, 0].set_title("Sine Function")

# For Cosine Function
axis[0, 1].plot(X, Y2)
axis[0, 1].set_title("Cosine Function")

# For Tangent Function
axis[1, 0].plot(X, Y3)
axis[1, 0].set_title("Tangent Function")

# For Tanh Function
axis[1, 1].plot(X, Y4)
axis[1, 1].set_title("Tanh Function")

# Combine all the operations and display
plt.suptitle("Trignometric Ratios", size=15)
plt.show()
```

In the same plot

```python
X = np.arange(0, math.pi*2, 0.05)

y = np.sin(X)
z = np.cos(X)

# Plotting both the curves simultaneously
plt.plot(X, y, color='r', label='sin')
plt.plot(X, z, color='g', label='cos')

# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine and Cosine functions")

# Adding legend, which helps us recognize the curve according to it's color
plt.legend()

# To load the display window
plt.show()
```

using add_subplot

```python
fig = plt.figure()
x = np.arange(0, 10, 0.01)
plt.plot(x, x-5)

plt.title("Not so Tiny")
fig.add_subplot(2, 2, 4)
plt.plot(x, 5-x)

plt.title("Tiny Plot")
plt.show()
```

Using add_axes

```python
fig = plt.figure(figsize =(5, 4))
plt.plot(np.sin(np.arange(0, 2*np.pi)))
# Creating a new axes for the figure
ax = fig.add_axes([0.5, 0.5, 0.3, 0.3])

# Adding the data to be plotted
ax.plot([2, 3, 4, 5, 5, 6, 6], [5, 7, 1, 3, 4, 6 ,8])
plt.show()
```
